```python
import random

def gieo_xuc_xac():
    x1 = random.randint(1, 6)
    x2 = random.randint(1, 6)
    tong = x1 + x2
    return x1, x2, tong

def phan_loai_tai_xiu(tong):
    if tong == 5:
        return "5"
    elif tong > 5:
        return "Tài"
    else:
        return "Xỉu"

def choi_game():
    tien = 100000
    so_van_thang = 0

    print("Chào mừng đến với trò chơi Tài Xỉu!")
    print("Bạn có 100,000 VNĐ để bắt đầu chơi.")

    while tien > 0:
        print(f"\n Số dư hiện tại: {tien:,} VNĐ")
        
        while True:
            try:
                tien_cuoc = int(input("Nhập số tiền bạn muốn cược (bội số của 1,000, tối thiểu 1,000): "))
                if 1000 <= tien_cuoc <= tien and tien_cuoc % 1000 == 0:
                    break
                else:
                    print("Số tiền cược không hợp lệ.")
            except ValueError:
                print("Vui lòng nhập số.")

        while True:
            lua_chon = input("Chọn (Tài / Xỉu / 5): ").strip().capitalize()
            if lua_chon in ["Tài", "Xỉu", "5"]:
                break
            print("Lựa chọn không hợp lệ. Vui lòng chọn Tài, Xỉu hoặc 5.")

        x1, x2, tong = gieo_xuc_xac()
        ket_qua = phan_loai_tai_xiu(tong)

        print(f"\n Xúc xắc: {x1} + {x2} = {tong} → {ket_qua}")

        if lua_chon == ket_qua:
            if ket_qua == "5":
                tien_thang = tien_cuoc * 3
                print(f"Thắng lớn! Bạn nhận được {tien_thang:,} VNĐ")
                tien += tien_thang
            else:
                print(f"Bạn thắng! Nhận {tien_cuoc:,} VNĐ")
                tien += tien_cuoc
            so_van_thang += 1
        else:
            print(f"Bạn thua! Mất {tien_cuoc:,} VNĐ")
            tien -= tien_cuoc

        if tien <= 0:
            print("\n Bạn đã hết tiền. Trò chơi kết thúc!")
            break

        choi_tiep = input("Bạn có muốn chơi tiếp không? (c/k): ").lower()
        if choi_tiep != 'c':
            break

    print("\n Thống kê kết quả:")
    print(f"- Số ván thắng: {so_van_thang}")
    print(f"- Số tiền còn lại: {tien:,} VNĐ")
    print("Cảm ơn bạn đã chơi!")

choi_game()
```
